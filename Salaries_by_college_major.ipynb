```python
#import pandas
import pandas as pd
```

```python
#read in csv file
df = pd.read_csv("salaries_by_college_major.csv")
```

```python
#look at first 5 rows of the dataset
df.head()
```

```python
#see number of rows
df.shape #51 rows and 6 columns
```

```python
#get access to the column names
df.columns
```

```python
#look at the last 5 rows
df.tail() #there's missing data
```

```python
#to get rid of the last row with missing values
clean_df = df.dropna()
```

```python
#looking at last five rows, there aren't missing values
clean_df.tail()
```

```python
#get access to a particular column, "Starting Median Salary"
clean_df["Starting Median Salary"]
```

```python
#to get the highest starting salary just add ".max()"
clean_df["Starting Median Salary"].max()
```

```python
#to get the index of the major that earns $74,300, we use ".idxmax()" to find the row
clean_df["Starting Median Salary"].idxmax() #turns out to be 43
```

```python
#to see the major associated with row 43, we use ".loc" for location with the column "Undergraduate Major"
clean_df["Undergraduate Major"].loc[43]
```

```python
#this is the second way to find the major but with just brackets and the row number
clean_df["Undergraduate Major"] [43]
```

```python
#to get the entire row, don't specify the column, but use ".loc"
clean_df.loc[43]
```

```python
#get highest mid-career salary
print(clean_df['Mid-Career Median Salary'].max())

#get the row/index where it's located
print(f"Index for the max mid career salary: {clean_df['Mid-Career Median Salary'].idxmax()}")

#get the major
clean_df['Undergraduate Major'][8]
```

```python
#get the lowest starting salary
print(clean_df['Starting Median Salary'].min())
clean_df['Undergraduate Major'].loc[clean_df['Starting Median Salary'].idxmin()]
```

```python
#get the lowest mid-career slary
clean_df.loc[clean_df['Mid-Career Median Salary'].idxmin()]
```

```python
#to get the difference between the 90th and 10th percentile salary of each major
clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
```

```python
#add the risk of each major's salary into the dataframe as a column using ".insert"
spread_col = clean_df['Mid-Career 90th Percentile Salary'] -clean_df['Mid-Career 10th Percentile Salary']

#naming the column "Spread"
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
#to find the major with the lowest risk use ".sort_values()" method
low_risk = clean_df.sort_values("Spread")

#just looking at the major and spread
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
#sort the salaries to highest earning potentials
highest_potential = clean_df.sort_values("Mid-Career 90th Percentile Salary", ascending = False)

#show the top five careers of earning potentials
highest_potential[["Undergraduate Major", "Mid-Career 90th Percentile Salary"]].head()
```

```python
#sort the careers by spread in decsending order
highest_spread = clean_df.sort_values("Spread", ascending = False)

#show the top five careers of the highest spread
highest_spread[["Undergraduate Major", "Spread"]].head()
```

```python
#getting the count of each major in the type of major, Business, HASS, and STEM
clean_df.groupby('Group').count()
```

```python
#get the average salary by group
pd.options.display.float_format = '{:,.2f}'.format
clean_df.groupby('Group').mean(numeric_only = True)
```
